```python
def graphs(x,y,x1,y1,col=["mediumorchid","olivedrab"],TITU=R"DCR vs Thershold para V=$71 V$",lim=[-10,200],lin=75, ECU="Thershold: 77mV"):
# Estilo
    plt.style.use('seaborn-v0_8-whitegrid')
    background_color = '#f8f9fa'
    plt.gca().set_facecolor(background_color)


    plt.plot(x, y, color=col[0], label="DCR con pastilla de Cesio")
    plt.plot(x1, y1, color=col[1], label="DCR sin pastilla de Cesio")
    plt.xlabel('Thershold [mV]', fontsize=13, fontweight='bold', color="#34495e")
    plt.ylabel('DCR [Hz]', fontsize=13, fontweight='bold', color="#34495e")
    plt.title(TITU,fontsize=16, fontweight='bold', pad=20, color="#2c3e50")
    plt.yscale('log')
    plt.xlim(lim[0],lim[1])

  # Cuadrícula
    plt.grid(True, alpha=0.8, linestyle='-', linewidth=0.5)
    
    # Línea vertical punteada entre Prueba 2 y Prueba 3
    plt.axvline(x=lin, color='red', linestyle='--', linewidth=2)

    #Anotación LaTeX
    ecuacion = ECU
    plt.annotate(ecuacion, xy=(0.2, 0.9), xycoords='axes fraction',
                fontsize=13, color="#2c3e50",
                bbox=dict(boxstyle="round,pad=0.3", facecolor="#ecf0f1", 
                edgecolor="#bdc3c7", alpha=0.9))


  # Ejes más limpios
    plt.gca().spines['top'].set_visible(False)
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['left'].set_color('#7f8c8d')
    plt.gca().spines['bottom'].set_color('#7f8c8d')

    plt.grid(axis='y', alpha=0.8)
    plt.legend(loc="best")
    plt.tight_layout()
    plt.show()
```

```python
 graphs(x,y,x1,y1,col=["mediumorchid","olivedrab"],TITU=R"DCR vs Thershold para V=$71 V$",lim=[-10,200],lin=75,ECU="Thershold: 75mV")
```
